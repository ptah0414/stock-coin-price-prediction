```python
import pyupbit
import numpy as np
import pandas as pd
```

```python
# OHLCV(open, high, low, close, volume)로 당일 시가, 고가, 저가, 종가, 거래량에 대한 데이터
df = pyupbit.get_ohlcv("KRW-BTC", count=365*10)
df
```

```python
df = df[["open", "high", "low", "close"]]
df
```

```python
# hold 누적 수익률: 현재 종가 / 매수 첫날 시가 
df['buy_hold_hpr'] = df['close'] / df.open[0]
df
```

```python
# Draw Down 계산 (누적 최대 값과 현재 hpr 차이 / 누적 최대값 * 100)
df['buy_hold_dd'] = (df['buy_hold_hpr'] - df['buy_hold_hpr'].cummax()) / df['buy_hold_hpr'].cummax() * 100
df
```

```python
df["buy_hold_MDD"] = df["buy_hold_dd"].cummin()
df
```

```python
k = 0.5
```

```python
# 변동폭 * k 계산, (고가 - 저가) * k값
df['range'] = (df['high'] - df['low']) * k
df
```

```python
# target(매수가), range 컬럼을 한칸씩 밑으로 내림(.shift(1))
df['target'] = df['open'] + df['range'].shift(1)
df
```

```python
# ror(수익률), np.where(조건문, 참일때 값, 거짓일때 값)
df['ror'] = np.where(df['high'] > df['target'],
                     df['close'] / df['target'],
                     1)
df
```

```python
# 누적 곱 계산(cumprod) => 누적 수익률
df['hpr'] = df['ror'].cumprod()
df
```

```python
# Draw Down 계산 (누적 최대 값과 현재 hpr 차이 / 누적 최대값 * 100)
df['dd'] = (df['hpr'] - df['hpr'].cummax()) / df['hpr'].cummax() * 100 
df
```

```python
# Draw Down 계산 (누적 최대 값과 현재 hpr 차이 / 누적 최대값 * 100)
df['MDD'] = df['dd'].cummin()
df
```

```python
df2 = pd.DataFrame()
```

```python
df2 = df[["hpr", "MDD", "buy_hold_hpr", "buy_hold_MDD"]]
df2
```

```python
df2["hpr_percent"] = (df2["hpr"] - 1) * 100
df2
```

```python
df2["buy_hold_hpr_percent"] = (df2["buy_hold_hpr"] - 1) * 100
df2
```
